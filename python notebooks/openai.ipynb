```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data - replace with actual values
data = {
    'Year': [2019, 2020, 2021, 2022, 2023],
    'Revenue': [100, 200, 400, 800, 1600],  # in million USD
    'Earnings': [10, 20, 50, 100, 200],    # in million USD
    'Gross Profit Margin': [20, 25, 30, 35, 40]  # in percentage
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot Revenue and Earnings
fig, ax1 = plt.subplots()

ax1.set_xlabel('Year')
ax1.set_ylabel('Revenue and Earnings (in million USD)')
ax1.plot(df['Year'], df['Revenue'], label='Revenue', color='tab:blue')
ax1.plot(df['Year'], df['Earnings'], label='Earnings', color='tab:green')
ax1.legend(loc='upper left')

# Format the x-axis to show only the year with no decimal points
ax1.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}'))

# Create a second y-axis to plot the Gross Profit Margin
ax2 = ax1.twinx()
ax2.set_ylabel('Gross Profit Margin (%)')
ax2.plot(df['Year'], df['Gross Profit Margin'], label='Gross Profit Margin', color='tab:red')
ax2.legend(loc='upper right')

plt.title('OpenAI Financial Performance Over the Past 5 Years')
plt.show()
```
