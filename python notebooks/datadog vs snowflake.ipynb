```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Example data to simulate stock prices (actual historical data should be sourced from financial platforms)
# Simulated stock price data for Datadog (DDOG) and Snowflake (SNOW) over the past 5 years

dates = pd.date_range(start="2019-01-01", periods=60, freq='M')

# Simulated stock price data (These are hypothetical values)
datadog_prices = np.cumsum(np.random.normal(loc=5, scale=10, size=len(dates))) + 40
snowflake_prices = np.cumsum(np.random.normal(loc=3, scale=8, size=len(dates))) + 60

# Creating a dataframe for easy plotting
df = pd.DataFrame({
    'Date': dates,
    'Datadog (DDOG)': datadog_prices,
    'Snowflake (SNOW)': snowflake_prices
})

# Plotting the stock prices
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Datadog (DDOG)'], label='Datadog (DDOG)', color='blue', linewidth=2)
plt.plot(df['Date'], df['Snowflake (SNOW)'], label='Snowflake (SNOW)', color='orange', linewidth=2)

plt.title('Datadog and Snowflake Stock Prices Over the Past 5 Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Stock Price (USD)', fontsize=12)
plt.legend()
plt.grid(True)

plt.xticks(rotation=45)
plt.tight_layout()

# Displaying the plot
plt.show()
```
