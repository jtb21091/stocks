```python
import pandas as pd

# Creating the dataframe
data = {
    "Year": [2023, 2022, 2021, 2020, 2019, 2018],
    "Free Cash Flow (in billions USD)": [7.0, 6.186, 2.516, -0.201, 4.497, 4.863]
}

df = pd.DataFrame(data)

# Saving to an Excel file
file_path = "Free_Cash_Flow.xlsx"
df.to_excel(file_path, index=False)

file_path
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading the Excel file
file_path = "Free_Cash_Flow.xlsx"
df_read = pd.read_excel(file_path)

# Display the dataframe (optional)
print(df_read)

# Analyzing and plotting the data
# Plotting the Free Cash Flow over the years
plt.figure(figsize=(10, 6))
plt.plot(df_read["Year"], df_read["Free Cash Flow (in billions USD)"], marker='o')
plt.title("Free Cash Flow Over Years")
plt.xlabel("Year")
plt.ylabel("Free Cash Flow (in billions USD)")
plt.grid(True)
plt.show()
```
