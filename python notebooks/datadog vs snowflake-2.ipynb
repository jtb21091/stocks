```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# Download historical stock data for Datadog (DDOG) and Snowflake (SNOW) over the past 5 years
datadog_data = yf.download('DDOG', start='2019-01-01', end='2024-01-01', interval='1mo')
snowflake_data = yf.download('SNOW', start='2019-01-01', end='2024-01-01', interval='1mo')

# Extracting the 'Close' prices for both stocks
datadog_prices = datadog_data['Close']
snowflake_prices = snowflake_data['Close']

# Aligning both datasets to have the same dates (inner join on date index)
df = pd.concat([datadog_prices, snowflake_prices], axis=1, join='inner')
df.columns = ['Datadog (DDOG)', 'Snowflake (SNOW)']
df.dropna(inplace=True)  # Drop any rows with missing values

# Plotting the stock prices
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Datadog (DDOG)'], label='Datadog (DDOG)', color='blue', linewidth=2)
plt.plot(df.index, df['Snowflake (SNOW)'], label='Snowflake (SNOW)', color='orange', linewidth=2)

plt.title('Datadog and Snowflake Stock Prices Over the Past 5 Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Stock Price (USD)', fontsize=12)
plt.legend()
plt.grid(True)

plt.xticks(rotation=45)
plt.tight_layout()

# Displaying the plot
plt.show()
```
