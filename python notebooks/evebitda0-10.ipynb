```python
import pandas as pd
import time

# Start time loader
start_time = time.time()

# Load the uploaded Excel file
file_path = 'sp500_ev_ebitda.xlsx'
data = pd.read_excel(file_path)

# Filter the dataframe for EV/EBITDA values between 0 and 10
filtered_data = data[(data['EV/EBITDA'] >= 0) & (data['EV/EBITDA'] <= 10)]

# Select only the relevant columns
filtered_data = filtered_data[['Company', 'Ticker', 'EV/EBITDA']]

# End time loader
end_time = time.time()
print(f"Execution Time: {end_time - start_time} seconds")

# Create a string representation of the filtered data
filtered_data_str = filtered_data.to_string(index=False)

# Print the string representation of the filtered data
print(filtered_data_str)

# Alternatively, if you want to use the custom tool to display the dataframe
import ace_tools as tools; tools.display_dataframe_to_user(name="Filtered Companies with EV/EBITDA between 0 and 10", dataframe=filtered_data)

filtered_data
```
